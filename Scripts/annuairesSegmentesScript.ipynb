```python
import pandas as pd

# Spécifiez le nom du fichier CSV
AnnuaireSegmente = "path2project/Annuaire/Annuaire_INSEE.xlsm"
               
# Utilisez la fonction read_excel() pour importer toutes les feuilles du fichier CSV
df = pd.read_excel(AnnuaireSegmente, sheet_name='EtablissementsModifies')
```

```python
df.shape
```

```python
df
```

```python
#drop each column for which all values are Nan
df.dropna(axis=1, how='all', inplace=True)
```

```python
#print first line
[[colname, value] for colname, value in zip(df.columns, df.iloc[0])]
```

```python
df.shape
```

```python
#save data to modify 
dh = df
```

```python
#Filtres
DenominationUniteLegale = ['PUBLIC', 'PUBLIQUE', 'NATIONAL', 'NATIONALE']
ActivitePrincipaleUniteLegaleSousClasse = ['Nettoyage', 'consulaires']
ActivitePrincipaleUniteLegaleSection = ['Agriculture', 'Administration', 'Enseignement']

```

```python
for i in DenominationUniteLegale:
    dh = dh[~(dh['denominationUniteLegale'].str.contains(i,case=False, na=False, regex=True))]

dh.shape
```

```python
for i in ActivitePrincipaleUniteLegaleSousClasse:
    dh = dh[~(dh['activitePrincipaleUniteLegale sous-classe'].str.contains(i, case=False, na=False, regex=True))]

dh.shape
```

```python
for i in ActivitePrincipaleUniteLegaleSection:
    dh = dh[~(dh['activitePrincipaleUniteLegale section'].str.contains(i,case=False, na=False, regex=True))]

dh.shape
```

```python
dh.to_csv('path2project/Annuaire/AnnuaireFiltre.csv', index=False, encoding='ANSI')
```

```python
#creation of segmented Anuaire_Insee 
# 500 et plus collaborateurs', '100 à 499 collaborateurs', '50 à 99 collaborateurs

Seg1 = dh[dh['trancheEffectifsEtablissement'].str.contains('50 à 99 salariés')]
Seg1.to_csv('path2project/Annuaire/AnnuaireSegmente/50 à 99 salariés.csv', index=False, encoding='ANSI')
```

```python
tranche1 = dh[dh['trancheEffectifsEtablissement'].str.contains('100 à 199 salariés')]

tranche2 = dh[dh['trancheEffectifsEtablissement'].str.contains('200 à 249 salariés')]

tranche3 = dh[dh['trancheEffectifsEtablissement'].str.contains('250 à 499 salariés')]

Seg2 = pd.concat([tranche1, tranche2, tranche3], axis=0)

Seg2.to_csv('path2project/Annuaire/AnnuaireSegmente/100 à 499 collaborateurs.csv', index=False, encoding='ANSI')
```

```python
tranche1 = dh[dh['trancheEffectifsEtablissement'].str.contains('10 000 salariés et plus')]

tranche2 = dh[dh['trancheEffectifsEtablissement'].str.contains('1000 à 999 salariés')]

tranche3 = dh[dh['trancheEffectifsEtablissement'].str.contains('2000 à 4999 salariés')]

tranche4 = dh[dh['trancheEffectifsEtablissement'].str.contains('500 à 499 salariés')]

tranche5 = dh[dh['trancheEffectifsEtablissement'].str.contains('5000 à 9999 salariés')]

Seg3 = pd.concat([tranche1, tranche2, tranche3, tranche4, tranche5], axis=0)

Seg3.to_csv('path2project/Annuaire/AnnuaireSegmente/500 collaborateurs et plus.csv', index=False, encoding='ANSI')
```
